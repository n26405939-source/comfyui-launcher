# Modular ComfyUI Launcher (Colab Edition)

**Strictly Ephemeral / GitHub Based**
1.  **Replace `REPO_URL` below** with your repository link.
2.  **Config Selection**: Choose between `z_image_project.json` (Custom UI) or `comfyui_server.json` (Standard ComfyUI).
3.  Run the cell.

```python
# ------------------------------------------------------------------------------
# NUCLEAR BOOTSTRAP (Self-Cleaning)
# ------------------------------------------------------------------------------
REPO_URL = "https://github.com/n26405939-source/comfyui-launcher.git"
REPO_NAME = "comfyui-launcher" 

import os, sys, shutil

def nuclear_reset():
    print("--- NUCLEAR RESET: INITIALIZING CLEAN ENVIRONMENT ---")
    
    # 1. Force Root Environment
    if os.path.exists('/content'):
        os.chdir('/content')
        print("Current Directory forced to: /content")
    
    # 2. Complete Cleanup (Prevent Nesting)
    if os.path.exists(REPO_NAME):
        print(f"Removing existing {REPO_NAME} to ensure fresh clone...")
        # Use system rm -rf for speed and reliability in Colab
        os.system(f"rm -rf {REPO_NAME}")
    
    # 3. Fresh Clone from the source
    print(f"Cloning {REPO_NAME} from GitHub...")
    clone_status = os.system(f"git clone {REPO_URL}")
    
    if clone_status == 0 and os.path.exists(REPO_NAME):
        os.chdir(REPO_NAME)
        sys.path.append(os.getcwd())
        print(f"Running from clean root: {os.getcwd()}")
        return True
    else:
        print("CRITICAL ERROR: Clone failed.")
        return False

if nuclear_reset():
    # --------------------------------------------------------------------------
    # CONFIGURATION
    # --------------------------------------------------------------------------
    
    # Options: templates/z_image_project.json | templates/comfyui_server.json
    CONFIG_PATH = "templates/comfyui_server.json" 

    print(f"Initializing Launcher for Config: {CONFIG_PATH}")

    try:
        from utils.comfy_launcher import ComfyLauncher
        launcher = ComfyLauncher(CONFIG_PATH) 
        launcher.launch()
    except Exception as e:
        import traceback
        print("\n--- LAUNCHER ERROR ---")
        traceback.print_exc()
```
