# Modular ComfyUI Launcher (Colab Edition)

**Strictly Ephemeral / GitHub Based**
1.  **Replace `REPO_URL` below** with your repository link.
2.  **Config Selection**: Choose between `z_image_project.json` (Custom UI) or `comfyui_server.json` (Standard ComfyUI).
3.  Run the cell.

```python
# ------------------------------------------------------------------------------
# BOOTSTRAP STRATEGY (Auto-Clone)
# ------------------------------------------------------------------------------
REPO_URL = "https://github.com/n26405939-source/comfyui-launcher.git"
REPO_NAME = "comfyui-launcher" 

import os
import sys
import shutil

# 1. Ensure we are in /content for a clean start on Colab
if os.path.exists('/content'):
    os.chdir('/content')
    # Clean up accidental nested folders if they exist from previous failed runs
    bad_nest = os.path.join(REPO_NAME, REPO_NAME)
    if os.path.exists(bad_nest):
        print("Cleaning up nested folders...")
        shutil.rmtree(bad_nest)

if not os.path.exists(REPO_NAME):
    print(f"Cloning {REPO_NAME} from GitHub...")
    os.system(f"git clone {REPO_URL}")

if os.path.exists(REPO_NAME):
    os.chdir(REPO_NAME)
    print("Pulling latest changes...")
    os.system("git pull")
    sys.path.append(os.getcwd())
    print(f"Running from: {os.getcwd()}")
else:
    print("Error: Clone failed or folder name mismatch.")

# ------------------------------------------------------------------------------
# CONFIGURATION
# ------------------------------------------------------------------------------

# Choose your project config:
CONFIG_PATH = "templates/comfyui_server.json" 

print(f"Initializing Launcher for Config: {CONFIG_PATH}")

if not os.path.exists(CONFIG_PATH):
    print(f"Error: Config not found at {CONFIG_PATH}.")
else:
    try:
        from utils.comfy_launcher import ComfyLauncher
        launcher = ComfyLauncher(CONFIG_PATH) 
        launcher.launch()
    except Exception as e:
        import traceback
        traceback.print_exc()
```
