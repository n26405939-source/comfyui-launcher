# Modular ComfyUI Launcher (Colab Edition)

**Strictly Ephemeral / GitHub Based**
1.  **Replace `REPO_URL` below** with your repository link.
2.  Run the cell.
3.  It will clone the repo, install everything, and launch the Z-Image-Turbo interface.
4.  **Enter your Prompts** in the Gradio UI that appears at the bottom.

```python
# ------------------------------------------------------------------------------
# BOOTSTRAP STRATEGY (Auto-Clone)
# ------------------------------------------------------------------------------
REPO_URL = "https://github.com/n26405939-source/comfyui-launcher.git"
REPO_NAME = "comfyui-launcher" # Name of the folder created by git clone

import os
import sys

if not os.path.exists(REPO_NAME):
    print(f"Cloning {REPO_NAME} from GitHub...")
    !git clone {REPO_URL}

if os.path.exists(REPO_NAME):
    os.chdir(REPO_NAME)
    print("Pulling latest changes...")
    !git pull
    sys.path.append(os.getcwd())
    print(f"Running from: {os.getcwd()}")
else:
    print("Error: Clone failed or folder name mismatch.")

# ------------------------------------------------------------------------------
# CONFIGURATION
# ------------------------------------------------------------------------------

# Path is relative to the repo root
CONFIG_PATH = "templates/z_image_project.json" 

print(f"Initializing Launcher for Config: {CONFIG_PATH}")

if not os.path.exists(CONFIG_PATH):
    print(f"Error: Config not found at {CONFIG_PATH}.")
else:
    try:
        from utils.comfy_launcher import ComfyLauncher
        # Launches into 'ComfyUI' folder in the current directory
        launcher = ComfyLauncher(CONFIG_PATH) 
        launcher.launch()
    except ImportError:
        print("Error: Could not import utils. Make sure you are in the repo root.")
    except Exception as e:
        import traceback
        traceback.print_exc()
```
