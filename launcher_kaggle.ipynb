# ComfyUI Launcher (Kaggle Edition)

**Ephemeral Models** - Models are downloaded to `/tmp` each session to save your 20GB persistent storage.

1. Enable GPU: Settings → Accelerator → GPU T4 x2
2. Run this cell
3. Look for the **PUBLIC URL** in the output

```python
# ------------------------------------------------------------------------------
# KAGGLE BOOTSTRAP
# ------------------------------------------------------------------------------
REPO_URL = "https://github.com/n26405939-source/comfyui-launcher.git"
REPO_NAME = "comfyui-launcher"

import os, sys, shutil

def kaggle_reset():
    print("--- KAGGLE LAUNCHER: INITIALIZING ---")
    
    # Force Kaggle working directory
    os.chdir('/kaggle/working')
    print(f"Working directory: {os.getcwd()}")
    
    # Clean and fresh clone
    if os.path.exists(REPO_NAME):
        print(f"Removing existing {REPO_NAME}...")
        shutil.rmtree(REPO_NAME)
    
    print(f"Cloning {REPO_NAME} from GitHub...")
    clone_status = os.system(f"git clone {REPO_URL}")
    
    if clone_status == 0 and os.path.exists(REPO_NAME):
        os.chdir(REPO_NAME)
        sys.path.append(os.getcwd())
        print(f"Running from: {os.getcwd()}")
        return True
    else:
        print("ERROR: Clone failed.")
        return False

if kaggle_reset():
    # --------------------------------------------------------------------------
    # CONFIGURATION (Kaggle-optimized: models in /tmp)
    # --------------------------------------------------------------------------
    CONFIG_PATH = "templates/z_image_kaggle.json"
    
    print(f"Config: {CONFIG_PATH}")
    print("Note: Models will be downloaded to /tmp (ephemeral storage)")
    
    try:
        from utils.comfy_launcher import ComfyLauncher
        launcher = ComfyLauncher(CONFIG_PATH)
        launcher.launch()
    except Exception as e:
        import traceback
        print("\n--- ERROR ---")
        traceback.print_exc()
```
